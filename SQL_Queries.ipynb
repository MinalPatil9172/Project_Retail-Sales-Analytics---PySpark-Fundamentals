## # STEP 3 - Register DataFrames as SQL views and run analytical queries

```python
#Calculate total revenue across all stores
sales_to_stores_df = spark.sql("""
select store_id,
    SUM(Quantity_int * price_int) AS total_revenue
from silver_db.cleand_sales_data
group by store_id
""")

display(sales_to_stores_df )
```

```python
sales_to_stores_df.write.format("delta").mode("overwrite").saveAsTable("gold_db.total_revenue_by_store")
```

```python
#Identify top-selling products by revenue
top_selling_product = spark.sql("""
select product_id, SUM(quantity_int * price_int) as total_revenue
from silver_db.cleand_sales_data
group by product_id
order by total_revenue DESC
limit 5  """)
```

```python
top_selling_product.write.format("delta").mode("overwrite").saveAsTable("gold_db.top_selling_products_by_revenue")
```

```python
#Compute daily sales totals
daily_sales_df = spark.sql("""
select sale_date, SUM(total_amount) as total_sales_daily
from silver_db.cleand_sales_data
group by sale_date
order by Sale_date Desc
""")

display(daily_sales_df )
```

```python
daily_sales_df.write.format("delta").mode("overwrite").saveAsTable("gold_db.daily_sales_totals")
```
