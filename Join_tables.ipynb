## STEP 4 - Perform multi-table joins to enrich sales data

```python
#Join enriched sales to stores: add store_name, city, state

sales_to_stores_df = spark.sql("select a.*,b.store_name,b.city,b.state from silver_db.cleand_sales_data a join silver_db.cleand_stores_data b on a.store_id=b.store_id")

display(sales_to_stores_df)
```

```python
# Save As Delta Table
sales_to_stores_df.write.format("delta").mode("overwrite").saveAsTable("gold_db.sales_to_stores")

```

```python
#Verify join correctness: check row counts

row_count_df = spark.sql("""
SELECT 'cleand_sales_data' AS table_name, COUNT(*) AS row_count
FROM silver_db.cleand_sales_data

UNION ALL

SELECT 'cleand_stores_data' AS table_name, COUNT(*) AS row_count
FROM silver_db.cleand_stores_data

""")

display(row_count_df)
```
